```python

def plotStats(xs, ys, yerr, label):
    plt.errorbar(xs, ys, yerr=yerr, label=label, elinewidth =0.5)
    
def showStats(title, ylabel, xlabel):
    plt.title(title)
    ax = plt.axes()        
    ax.yaxis.grid()
    ax.set_ylabel(ylabel)
    ax.set_xlabel(xlabel)
    plt.legend()
    plt.show()
```
